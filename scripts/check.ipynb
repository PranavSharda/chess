```python
import httpx

async def verify_user(username: str) -> httpx.Response:
    async with httpx.AsyncClient() as client:
        response = await client.get(
            f"https://api.chess.com/pub/player/vialize/games/archives",
            timeout=10000.0
        )
    return response
```

```python
import json

response = await verify_user('chess1234578')
data = response.json()

# Save to file
with open('response_data.json', 'w') as f:
    json.dump(data, f, indent=2)

print(f"Response saved to response_data.json")
```
